```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
df= pd.read_excel('/content/drive/MyDrive/8 semestre/IA 2024/DatosRegLineal.xlsx')
```

```python
df.info()
```

```python
df.drop(axis=0, index=30, inplace=True)
df.drop(axis=0, index=31, inplace=True)
df.info()
```

```python
X=df['Kg. año']
Y=df['Costo anual']
```

```python
plt.scatter(X,Y)
plt.xlabel('Kg. año')
plt.ylabel('Costo anual')
plt.show()
```

```python

modelo=LinearRegression(fit_intercept=True)
modelo.fit(X.values.reshape(-1,1),Y.values.reshape(-1,1))
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X, Y, color='red')
plt.title('Regresion gradiente-descendiente')
plt.xlabel('Kg. año')
plt.ylabel('Costo anual')
plt.plot(X.values.reshape(-1,1), modelo.predict(X.values.reshape(-1,1)), color='blue')
plt.show()
```

```python
theta0 = modelo.intercept_
theta1 = modelo.coef_[0]
print(f'h(x)= {theta0} + {theta1}x')
```

```python
X_datosp=[45000,91000,125000,130000,150000,109000,100000,70000,110000,5000]
Y_datosp=[200000,450000,530000,540000,570000,470000,450000,300000,490000,70000]
```

```python
modelo.predict(np.array(X_datosp).reshape(-1,1))
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X_datosp, Y_datosp, color='red')
plt.title('Regresion gradiente-descendiente')
plt.xlabel('Kg. año')
plt.ylabel('Costo anual')
plt.plot(np.array(X_datosp).reshape(-1,1), modelo.predict(np.array(X_datosp).reshape(-1,1)), color='blue')
plt.show()
modelo.score(np.array(X_datosp).reshape(-1,1), np.array(Y_datosp).reshape(-1,1))
```
